```python
%matplotlib inline
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('lena-465x465.jpg',0)
plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```
