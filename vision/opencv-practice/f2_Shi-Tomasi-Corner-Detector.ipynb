```python
%matplotlib inline
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

# Shi-Tomasi Corner Detector & Good Features to Track Goal
https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_feature2d/py_shi_tomasi/py_shi_tomasi.html#shi-tomasi

* We will learn about the another corner detector: Shi-Tomasi Corner Detector
* We will see the function: cv2.goodFeaturesToTrack()

```python
img = cv2.imread('home.jpg')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

corners = cv2.goodFeaturesToTrack(gray,25,0.01,10)
corners = np.int0(corners)

for i in corners:
    x,y = i.ravel()
    cv2.circle(img,(x,y),3,255,-1)

plt.imshow(img),plt.show()
```
